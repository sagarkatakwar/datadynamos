```python
import pandas as pd

# Load your dataset
df = pd.read_csv("/content/Weather_london_preprocess.csv")

# Split and extract components from the 'date' column (assuming date format is consistent)
df[['day', 'month', 'date', 'year', 'time_range']] = df['date'].str.extract(
    r'^(.*?)(?:\s+)(\w+)(?:\s+)(\d+),\s+(\d{4}),\s+(.*)$'
)

# Clean up the extracted 'month' and 'date' columns
df['month'] = df['month'].str.strip()
df['date'] = df['date'].astype(int)

# Remove the comma in the 'day' column
df['day'] = df['day'].str.replace(',', '', regex=False)

# Remove the '+' sign from the 'date_id' column
df['date_id'] = df['date_id'].astype(str).str.replace('+', '', regex=False)

# Rearrange the columns for easier reading
df = df[['day', 'month', 'date', 'year', 'time_range', 'date_id', 'wind', 'desc', 'temp', 'baro', 'hum']]


# Output the first few rows to verify the changes
print(df.head())
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the cleaned dataset
df = pd.read_csv("/content/Weather_london_cleaned.csv")

# Plot temperature (°F) over time ranges using a line plot
plt.figure(figsize=(12, 6))
sns.lineplot(x='time_range', y='temp', data=df, marker='o', color='b')
plt.title('Temperature Over Time Ranges')
plt.xlabel('Time Range')
plt.ylabel('Temperature (°F)')
plt.xticks(rotation=45)
plt.tight_layout()

# Show plot
plt.show()
```

```python
# Plot wind speed across time ranges using a bar plot
plt.figure(figsize=(12, 6))
sns.barplot(x='time_range', y='wind', data=df, palette="Blues")
plt.title('Wind Speed Across Time Ranges')
plt.xlabel('Time Range')
plt.ylabel('Wind Speed (m/s)')
plt.xticks(rotation=45)
plt.tight_layout()

# Show plot
plt.show()
```

```python
# Plot humidity levels across time ranges using a bar plot
plt.figure(figsize=(12, 6))
sns.barplot(x='time_range', y='hum', data=df, palette="Greens")
plt.title('Humidity Levels Across Time Ranges')
plt.xlabel('Time Range')
plt.ylabel('Humidity (%)')
plt.xticks(rotation=45)
plt.tight_layout()

# Show plot
plt.show()
```

```python
# Plot a histogram to show the distribution of wind speeds
plt.figure(figsize=(10, 6))
sns.histplot(df['wind'], kde=True, color='red', bins=15)
plt.title('Distribution of Wind Speed')
plt.xlabel('Wind Speed (m/s)')
plt.ylabel('Frequency')
plt.tight_layout()

# Show plot
plt.show()
```

```python
# Plot a histogram for temperature distribution (°F)
plt.figure(figsize=(10, 6))
sns.histplot(df['temp'], kde=True, color='orange', bins=15)
plt.title('Temperature Distribution')
plt.xlabel('Temperature (°F)')
plt.ylabel('Frequency')
plt.tight_layout()

# Show plot
plt.show()
```

```python
# Calculate correlations between numeric columns and visualize using a heatmap
plt.figure(figsize=(10, 6))
corr_matrix = df[['wind', 'temp', 'baro', 'hum']].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.tight_layout()

# Show plot
plt.show()
```

```python
# Plot the average temperature for each day of the week using a bar plot
average_temp = df.groupby('day')['temp'].mean().reset_index()
plt.figure(figsize=(12, 6))
sns.barplot(x='day', y='temp', data=average_temp, palette='coolwarm')
plt.title('Average Temperature for Each Day of the Week')
plt.xlabel('Day')
plt.ylabel('Average Temperature (°F)')
plt.tight_layout()

# Show plot
plt.show()
```

```python
# Scatter plot to show the relationship between temperature and wind speed
plt.figure(figsize=(12, 6))
sns.scatterplot(x='temp', y='wind', data=df, hue='day', palette='tab10')
plt.title('Temperature vs Wind Speed')
plt.xlabel('Temperature (°F)')
plt.ylabel('Wind Speed (m/s)')
plt.tight_layout()

# Show plot
plt.show()
```

```python
# Scatter plot to show the relationship between temperature and humidity
plt.figure(figsize=(12, 6))
sns.scatterplot(x='temp', y='hum', data=df, hue='day', palette='muted')
plt.title('Temperature vs Humidity')
plt.xlabel('Temperature (°F)')
plt.ylabel('Humidity (%)')
plt.tight_layout()

# Show plot
plt.show()
```

```python
# Box plot to show the distribution of temperature across days
plt.figure(figsize=(12, 6))
sns.boxplot(x='day', y='temp', data=df, palette='Set2')
plt.title('Temperature Distribution Across Days')
plt.xlabel('Day')
plt.ylabel('Temperature (°F)')
plt.tight_layout()

# Show plot
plt.show()
```

```python
# Box plot showing how the temperature varies across different time ranges
plt.figure(figsize=(12, 6))
sns.boxplot(x='time_range', y='temp', data=df, palette='viridis')
plt.title('Temperature Distribution Across Time of Day')
plt.xlabel('Time Range')
plt.ylabel('Temperature (°F)')
plt.xticks(rotation=45)
plt.tight_layout()

# Show plot
plt.show()
```

```python
# Plot a histogram for the distribution of humidity levels
plt.figure(figsize=(10, 6))
sns.histplot(df['hum'], kde=True, color='purple', bins=15)
plt.title('Humidity Distribution')
plt.xlabel('Humidity (%)')
plt.ylabel('Frequency')
plt.tight_layout()

# Show plot
plt.show()
```

```python
# Plot the average wind speed by day of the week
average_wind = df.groupby('day')['wind'].mean().reset_index()
plt.figure(figsize=(12, 6))
sns.barplot(x='day', y='wind', data=average_wind, palette='Blues_d')
plt.title('Average Wind Speed by Day')
plt.xlabel('Day')
plt.ylabel('Average Wind Speed (m/s)')
plt.tight_layout()

# Show plot
plt.show()
```

```python
# Scatter plot to show the relationship between temperature and barometric pressure
plt.figure(figsize=(12, 6))
sns.scatterplot(x='temp', y='baro', data=df, hue='month', palette='coolwarm')
plt.title('Temperature vs Barometric Pressure')
plt.xlabel('Temperature (°F)')
plt.ylabel('Barometric Pressure (inHg)')
plt.tight_layout()

# Show plot
plt.show()
```

```python
# Plot hourly temperature changes by converting 'time_range' to numerical hours
df['hour'] = df['time_range'].str.extract(r'(\d{1,2})').astype(int)

plt.figure(figsize=(12, 6))
sns.lineplot(x='hour', y='temp', data=df, marker='o', color='green')
plt.title('Hourly Temperature Changes')
plt.xlabel('Hour of Day')
plt.ylabel('Temperature (°F)')
plt.tight_layout()

# Show plot
plt.show()
```

```python
# Scatter plot showing the relationship between wind speed and barometric pressure
plt.figure(figsize=(12, 6))
sns.scatterplot(x='wind', y='baro', data=df, hue='day', palette='Set1')
plt.title('Wind Speed vs Barometric Pressure')
plt.xlabel('Wind Speed (m/s)')
plt.ylabel('Barometric Pressure (inHg)')
plt.tight_layout()

# Show plot
plt.show()
```

```python
# Calculate the average temperature and wind speed across days and plot it
average_temp_wind = df.groupby('day')[['temp', 'wind']].mean().reset_index()
average_temp_wind = pd.melt(average_temp_wind, id_vars=['day'], value_vars=['temp', 'wind'], var_name='Measure', value_name='Value')

plt.figure(figsize=(12, 6))
sns.barplot(x='day', y='Value', hue='Measure', data=average_temp_wind, palette='Paired')
plt.title('Average Temperature and Wind Speed Across Days')
plt.xlabel('Day')
plt.ylabel('Value')
plt.tight_layout()

# Show plot
plt.show()
```

```python
# Pivot the data for a heatmap of temperature variation across days and time of day
temp_pivot = df.pivot_table(index='time_range', columns='day', values='temp', aggfunc='mean')

plt.figure(figsize=(12, 6))
sns.heatmap(temp_pivot, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Temperature Variation Across Days and Time of Day')
plt.xlabel('Day')
plt.ylabel('Time Range')
plt.tight_layout()

# Show plot
plt.show()
```

```python
# Violin plot to show the distribution of wind speed across days
plt.figure(figsize=(12, 6))
sns.violinplot(x='day', y='wind', data=df, palette='Set3')
plt.title('Wind Speed Distribution Across Days')
plt.xlabel('Day')
plt.ylabel('Wind Speed (m/s)')
plt.tight_layout()

# Show plot
plt.show()
```
